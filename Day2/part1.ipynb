```python
def repeated_ids_in_range(start, end):
    total = 0
    start_len = len(str(start))
    end_len = len(str(end))
    
    # Generate all repeated IDs by length of half the digits
    for length in range((start_len + 1) // 2, (end_len // 2) + 1):
        low = 10 ** (length - 1)
        high = 10 ** length - 1
        for half_num in range(low, high + 1):
            rep_str = str(half_num) * 2
            rep_num = int(rep_str)
            if rep_num > end:
                break
            if rep_num >= start:
                total += rep_num
    return total

def main():
    with open("input1.txt", "r") as file:
        ranges_line = file.read().strip()
    
    total_invalid = 0
    ranges = ranges_line.split(',')
    
    for r in ranges:
        if '-' not in r:
            continue
        start_str, end_str = r.split('-')
        start = int(start_str)
        end = int(end_str)
        total_invalid += repeated_ids_in_range(start, end)
    
    with open("output1.txt", "w") as outfile:
        outfile.write(str(total_invalid) + "\n")

if __name__ == "__main__":
    main()
```
