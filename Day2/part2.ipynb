```python
def is_invalid_id(num_str):
    """Check if number is made of some sequence repeated >=2 times"""
    n = len(num_str)
    
    # Check all possible sequence lengths that divide the number
    for seq_len in range(1, n // 2 + 1):
        if n % seq_len == 0:
            repeats = n // seq_len
            if repeats >= 2:
                seq = num_str[:seq_len]
                if seq * repeats == num_str:
                    return True
    return False

def main():
    total_invalid = 0
    
    with open("input2.txt", "r") as file:
        ranges_line = file.read().strip()
    
    ranges = ranges_line.split(',')
    for range_str in ranges:
        if '-' not in range_str:
            continue
        start_str, end_str = range_str.split('-')
        start = int(start_str)
        end = int(end_str)
        
        # Check each ID in range
        for num in range(start, end + 1):
            num_str = str(num)
            if is_invalid_id(num_str):
                total_invalid += num
    
    with open("output2.txt", "w") as outfile:
        outfile.write(str(total_invalid) + "\n")


if __name__ == "__main__":
    main()
```
